##### Google Ads

```python
#Set locale to Spanish
locale.setlocale(locale.LC_TIME, 'es_ES.UTF-8')

google_ads = pd.read_csv('/Users/cesarchaparro/Desktop/GitHub/Chapoteos/analytics_data/analytics/14-27.04.2025/csv/google_ads.csv', encoding='utf-8')
# Convert 'Día' to datetime, handling potential errors
google_ads['Día'] = pd.to_datetime(google_ads['Día'], errors='coerce')
# Convert 'Impr.' to integer
google_ads['Impr.'] = google_ads['Impr.'].astype('int64')
# Convert CTR to floatin and delete the % symbol
google_ads['CTR'] = google_ads['CTR'].str.replace(',', '', regex=False).str.rstrip('%').astype(float).div(100)
# Convert 'Coste' to float
google_ads['Coste'] = google_ads['Coste'].str.replace(',', '', regex=False).astype(float)
# Convert 'Conversiones' to integer
google_ads['Conversiones']  = google_ads['Conversiones'].str.replace(',', '', regex=False).astype('int64')
# Convert 'Ingresos' to float
google_ads['Ingresos'] = google_ads['Ingresos'].str.replace(',', '', regex=False).astype(float)
google_ads.head()
```
